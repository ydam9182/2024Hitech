```python
import pandas as pd
import tensorflow as tf
```

```python
# 파일들로부터 데이터 읽어오기

파일경로 = 'https://raw.githubusercontent.com/ydam9182/2024Hitech/main/lemonade.csv'
lemonade = pd.read_csv(파일경로)
```

```python
# 변수안에 어떤 데이터가 있는지 확인(데이터 모양으로 확인)
print(lemonade.shape)
```

```python
# 칼럼 이름 출력
print(lemonade.columns)
```

```python
x = lemonade[['온도']]
y = lemonade[['판매량']]
print(x.shape, y.shape)
```

```python
lemonade.head()
```

```python
X = tf.keras.layers.Input(shape=[1])
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X,Y)
model.compile(loss='mse')
```

```python
model.fit(x,y, epochs=1000)
```

```python
model.predict([[15]])
```
